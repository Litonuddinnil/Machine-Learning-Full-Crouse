```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = load_breast_cancer(return_X_y=True)
print('Feature matrix shape:', X.shape)
print('Target vector shape:', y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
print('Training samples:', X_train.shape[0])
print('Testing samples:', X_test.shape[0])
```

```python
stump = DecisionTreeClassifier(max_depth=1,random_state=42) 
stump.fit(X_train,y_train) 
Y_predict_stump = stump.predict(X_test) 
print("Accuracy: ",accuracy_score(y_test,Y_predict_stump))
                                
```

```python
base_learner = DecisionTreeClassifier(random_state=42,max_depth=1) 
ada = AdaBoostClassifier(
    estimator=base_learner,learning_rate=0.3,n_estimators=25,random_state=42) 
ada.fit(X_train,y_train)

y_predict_ada = ada.predict(X_test) 
print("Accuracy: ",accuracy_score(y_test,y_predict_ada))
```

```python
 

 
# Random Forest model
rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=None,
    random_state=42
)

# Train
rf.fit(X_train, y_train)

# Predict
y_pred = rf.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
